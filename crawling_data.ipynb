In this notebook we crawl data from www.psychologytoday.com. 
We store articles that were published on this site from 2015 till today.

```python
from bs4 import BeautifulSoup
import urllib.request
from datetime import datetime
import pandas as pd
import time
```

```python
# First page of blog posts.
page_number = 871
starting_url = f'https://www.psychologytoday.com/intl/blog-posts?page={page_number}'
```

```python
# Set user agent
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'

# this will be checked to see when to stop. 
crawler_date = datetime.now()
```

```python
temp_l = []
global temp_df

while crawler_date.year > 2015:

#     # Development settings.
#     if page_number > 2:
#         break
        
    print('Pagenumber:'+str(page_number))
    starting_url = f'https://www.psychologytoday.com/intl/blog-posts?page={page_number}'
    
    # Send request.
    request = urllib.request.Request(starting_url,headers={'User-Agent': user_agent})
    response = urllib.request.urlopen(request)

    # Page content.
    page_content = BeautifulSoup(response, "html.parser")

    # Extract teasers for articles.
    all_article_teasers = page_content.findAll("div", {"class": "blog-entry"})

    # The articles will be stored in this DataFrame.
    temp_df = pd.DataFrame({}, columns=['category', 'title', 'date', 'author', 'teaser_text', 'blog_post_link'])

    # Extract properties from teasers.
    for article in all_article_teasers: 
        temp_author_publish_date = article.find('p', 'blog_entry__publish_info').text[3:].split(' on ')
        date = datetime.strptime(temp_author_publish_date[1].split(' in ')[0], '%B %d, %Y')    
        category = article.find('h6').text
        title = article.find('h2').text
        author = temp_author_publish_date[0]
        blog_post_link = article.find('h2').find('a', href=True)['href']
        teaser_text = article.find('p', {'class':['blog_entry__teaser', 'd-sm-none']}).text
        
        try:
            # Extract the article itself.
            url = f'https://www.psychologytoday.com/{blog_post_link}'
            print(url)

            # Get article content.
            article_request = urllib.request.Request(url,headers={'User-Agent': user_agent})
            article_response = urllib.request.urlopen(article_request)
            article_page_content = BeautifulSoup(article_response, "html.parser")

            # Get div in which article is stored.
            article_div = article_page_content.find_all('div', {'class':['field', 'field-name-body', 'field-type-text-with-summary', 'field-label-hidden']})
            article_div = str(article_div)

            # Add everything to a list.
            temp_l.append([date, str(category), str(title), str(author), str(blog_post_link), str(teaser_text), article_div, f'/intl/blog-posts?page={page_number}', str(datetime.now())])
        except:
            print('Didn\'t find article')
                  
    # Store articles every 30 pages.
    if (page_number % 30 == 0) and (page_number > 0):
        temp_df = pd.DataFrame(temp_l, columns=['date', 'category', 'title', 'author', 'blog_post_link', 'teaser_text', 'article_div', 'crawled_from', 'time_crawled'])
        print('Stored articles')            
        temp_df.to_pickle(f'data/PT_data_{page_number}.pkl')
        temp_df = []
        # Reset List.
        temp_l = []
    
    # Next page.    
    page_number += 1
print('Done')
```

```python
temp_df = pd.DataFrame(temp_l, columns=['date', 'category', 'title', 'author', 'blog_post_link', 'teaser_text', 'article_div', 'crawled_from', 'time_crawled'])
print('Stored articles')            
temp_df[temp_df.date >= '01-01-2016'].to_pickle(f'data/PT_data_last.pkl') 
```
